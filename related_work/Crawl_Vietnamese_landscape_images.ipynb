# Upload txt file contain urls and clean it

```python
def clean_url_file(file_name):
    lines = []
    with open(file_name, "r") as f:
        lines = f.readlines()
        print("num of lines before process=", len(lines))
        lines = set(lines)
        f.close()

    with open(file_name,"r+") as f:
        f.seek(0)
        f.truncate()
        f.close()

    with open(file_name, "w") as f:
        for line in lines:
            f.write(line)
        print("num of lines after process=", len(lines))
        f.close()
```

```python
clean_url_file(txt_urls_filename)
```

# Crawl HR images from urls


```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from PIL import Image
import io
import os

def crawl_image_urls_from_a_page(pageTarget, min_width, saved_dir, image_name_state):
    print(pageTarget)
    page = requests.get(pageTarget)
    soup = BeautifulSoup(page.content, 'html.parser')
    wrapper = soup.find('body')
    images = wrapper.find_all("img")
    for image in images:
        try:
            image_url = image['src']
            if '.svg' in image_url: continue
            response = requests.get(image_url)
            image = Image.open(io.BytesIO(response.content))
            width, height = image.size
            if width < min_width: continue

            index = image_url.find("?")
            if index != -1:
                image_url = image_url[:index]
            print(image_url, image.size)

            # Lưu ảnh vào thư mục (mục tiêu là 1000 ảnh -> đặt tên ảnh gồm 4 số)
            image_name_state += 1
            image.save(os.path.join(saved_dir, f"{image_name_state:04}.png"))
            print(os.path.join(saved_dir, f"{image_name_state:04}.png"), "saved")
        except: pass
    return image_name_state


def crawl_and_save_images(url_pages, saved_dir, min_width, image_name_state):
    if not os.path.exists(os.path.join(saved_dir)):
        os.mkdir(os.path.join(saved_dir))

    # Cào tất cả các url hình ảnh đạt tiêu chuẩn
    with open(url_pages, "r") as f:
        lines = f.readlines()
        cnt = 0
        print("num of urls =", len(lines))
        for line in lines:
            cnt += 1
            print(f"url {cnt}")
            line = line.strip()
            image_name_state = crawl_image_urls_from_a_page(line, min_width, saved_dir, image_name_state)
            print("Saving the images is done!")

    print("Crawl is done!")

```
